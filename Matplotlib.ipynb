```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

Tutorial link = https://github.com/matplotlib/matplotlib/blob/main/tutorials/introductory/lifecycle.py

```python
data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
```

```python
group_data
```

```python
group_names
```

```python
#customize style
plt.style.use('seaborn')
#create figure and ajust size
fig, ax = plt.subplots(figsize=(10,5))
#plot
ax.barh(group_names, group_data)
# rotate labels
labels = ax.get_xticklabels()
#set property of many items
plt.setp(labels, rotation=45,horizontalalignment="right")
#add labels
ax.set(title='Company Revenue', xlabel = 'Revenue', ylabel= 'Company');
```

```python
fig.savefig('practice.png')
```

```python
#customize style
plt.style.use('seaborn')
#create figure and ajust size
fig, ax = plt.subplots(figsize=(10,5))
#plot
ax.barh(group_names, group_data)
# rotate labels
labels = ax.get_xticklabels()
#set property of many items
plt.setp(labels, rotation=45,horizontalalignment="right")
#add labels
ax.set(title='Company Revenue', xlabel = 'Revenue ($)', ylabel= 'Company')
#add vertical line
ax.axvline(group_mean,ls='--',color='r');
```

```python
fig.savefig('practice2.png')
```
